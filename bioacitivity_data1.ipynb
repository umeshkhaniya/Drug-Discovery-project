Importing packages

```python
from chembl_webresource_client.new_client import new_client
import pandas as pd
```

Target serach for Aromatase Inhibitor

```python
# Target serach for Aromatase Inhibitor
target = new_client.target
target_query = target.search('Aromatase inhibitor')
targets = pd.DataFrame.from_dict(target_query)
targets
```

Select the first entry which is single protein and target organism that drug bind is Homo sapiens

```python
selected_target = targets.target_chembl_id[0]
selected_target
```

Filter based on IC50 so that it will be same unit

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
df.shape
```

Drug potency depends on the standrd_value. Less the standard_value better potential of  drug.
If any compound has any missing data values for standard_value then need to drop that compound

```python
df1 = df[df.standard_value.notna()]
df1
```

Labeling the compunds based in IC50 unit (standard value). Compounds having values of less than 1000 nM will be considered to be active while those greater than 10,000 nM will be considered to be inactive. As for those values in between 1,000 and 10,000 nM will be referred to as intermediate.

We only need 3 columns molecule_chembl_id, canonical_smiles and standard_value. Drop others column

```python
df2 = df1[["molecule_chembl_id", "canonical_smiles", "standard_value"]].reset_index(drop = True)
df2
```

Add one column for activity based on standard_value

```python
# assign inactive, intermediate and active
bioactivity_class = []
for i in df2.standard_value:
    if float(i) >= 10000:
        bioactivity_class.append("inactive")
    elif float(i) <= 1000:
        bioactivity_class.append("active")
    else:
        bioactivity_class.append("intermediate")
bioactivity_class = pd.Series(bioactivity_class, name = "class")
bioactivity_class 
```

```python
# Let cancatinate the bioactivity data frame
df3 = pd.concat([df2, bioactivity_class], axis = 1)
df3
```

```python
# we only need few column here: molecule chemble id, smiles string and standard value
df3 = df2[['molecule_chembl_id', 'canonical_smiles', 'standard_value']]
df3
```

```python
# let concatinate the bioactivity data frame
df4 = pd.concat([df3, pd.Series(bioactivity_class)], axis = 1)
df4
```

```python
# save the data in csv file
df4.to_csv('bioactivity_preprocessed_data.csv', index=False)
```
